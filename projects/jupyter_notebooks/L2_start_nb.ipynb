## Set up steps

```python
import aerospike
import pandas
from itables import show

config = {'hosts': [('localhost',3000)]}

client = aerospike.client(config).connect()

ns = 'vest_vault'
sn = 'inventory'
```

## Item List

```python
item_ids = [1001,1002,1003,1004,1010,1011]
keyTuplesList = []

def create_keyTuple(ns, sn, ii):
    keyTuple = (ns, sn, ii)
    keyTuplesList.append(keyTuple)

for item_id in item_ids:
    create_keyTuple(ns,sn,item_id)
  
print(keyTuplesList)
```
